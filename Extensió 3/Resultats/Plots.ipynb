```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Experiments - Predecessors.csv')

plt.figure(figsize=(10, 6))
plt.plot(df['NÂº Books'], df['Time (ms)'], marker='o', linestyle='-')

plt.title('Time vs Number of Books')
plt.xlabel('Number of Books')
plt.ylabel('Time Taken (ms)')

plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Experiments - Predecessors.csv')

plt.figure(figsize=(10, 6))
plt.plot(df['NÂº Books'], df['Time (ms)'], marker='o', linestyle='-')

plt.title('Time vs Number of Books (Log Scale on Y-axis)')
plt.xlabel('Number of Books')
plt.ylabel('Time Taken (ms)')

plt.yscale('log')

plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Experiments - Predecessors.csv')

plt.figure(figsize=(10, 6))
plt.plot(df['NÂº Books'], int(df['Time (ms)'])**2, marker='o', linestyle='-')

plt.title('Time vs Number of Books (Quadratic Scale on Y-axis)')
plt.xlabel('Number of Books')
plt.ylabel('Time Taken (ms)^2')  # Mostrar que se aplicÃ³ la escala cuadrÃ¡tica

plt.grid(True)
plt.show()
```
